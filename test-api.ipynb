```python
import requests

# Базовый URL (адрес скрипта)
url = 'https://superbank.tryapi.ru/clients/'

# Параметры запроса (то, что идет после знака вопроса в URL)
params = {
    'page': 1,
    'limit': 5
}

try:
    # Выполняем GET-запрос
    response = requests.get(url, params=params)

    # Проверяем, что запрос прошел успешно (код 200)
    if response.status_code == 200:
        # Получаем данные в формате JSON
        data = response.json()
        
        # Выводим результат
        # print("Данные получены успешно:")
        # print(data)
        
        # Если нужно посмотреть конкретные поля, можно сделать так:
        for client in data['data']:
            print(client.get('fullname'), "-", client.get("birthday"), "-", client.get("phone"))
            # print(client)
    else:
        print(f"Ошибка сервера: код {response.status_code}")

except requests.exceptions.RequestException as e:
    print(f"Ошибка подключения: {e}")
```
